```python
# 2.  Average Scores by Gender (Hint : Use bar chart for graph and use pd.groupbyÂ forÂ gender)
#  This is the Chatgpt code
avg_by_gender = df.groupby("gender")["average"].mean().reset_index()

# Plot bar chart
sns.barplot(x="gender", y="average", data=avg_by_gender)

# Show chart
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xlabel("Gender")
plt.show()
```
