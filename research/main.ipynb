```python
#  Sir code
def getCsv(file):
    df = pd.read_csv(file)
    return df

# getCsv()
df = getCsv("E:/CEDMAP python/Code Sarvarth/Project1/data/raw.csv")
df
```

```python
all_scores = pd.melt(df, id_vars=[],value_vars=["math_score", "reading_score", "writing_score"],var_name="subject", value_name="score")
```

```python
all_scores["subject"] = all_scores["subject"].str.replace("_score", "")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=all_scores, x="score", hue="subject", multiple="stack", bins=10)
plt.title("Histogram of Scores")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()
```

```python
avg_scores = df.groupby("gender")[["math_score", "reading_score", "writing_score"]].mean().reset_index()
```

```python
avg_scores_melted = pd.melt(avg_scores, id_vars="gender", var_name="subject", value_name="average_score")
avg_scores_melted["subject"] = avg_scores_melted["subject"].str.replace("_score", "")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=avg_scores_melted, x="subject", y="average_score", hue="gender")
plt.title("Average Scores by Gender")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.legend(title="Gender")
plt.show()
```
